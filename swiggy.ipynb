```python
#Price vs. estimated revenue
plt.figure(figsize=(8,5))
sns.regplot(data=df,x='Price',y='Estimated_revenue',scatter_kws={'alpha':0.4, 's':20},line_kws={'color':'red'})
plt.title("Price vs Estimated Revenue")
plt.xlabel("Price (INR)")
plt.ylabel("Estimated Revenue")
plt.xlim(0, df['Price'].quantile(0.99))
plt.ylim(0, df['Estimated_revenue'].quantile(0.99))
plt.show()
```

1. Correlation Analysis (Regression Line)

The regression line shows a clear positive correlation :
- higher-priced items are associated with higher estimated revenue.
- The slope is meaningfully upward, indicating that pricing directly contributes to revenue potential.
- The confidence band is narrow at lower prices but widens at higher prices, which signals greater variability in revenue at premium price levels.

2. Distribution Analysis (Critical Observation) : The scatter plot reveals something highly unusual and non-random:

- Distinct Linear Clusters :
The points are arranged in multiple straight-line clusters, not a natural scatter. These lines originate near zero but rise at different slopes.
- This pattern strongly indicates that Estimated Revenue was not an observed value — it was likely generated using specific formulas or tiers, for example:

Revenue = Price x constant

Different constants for different groups → multiple straight lines.

3. Implications for the Project :

Business Insight :
- There is a positive relationship:
- Higher price → higher potential revenue.
- That part is legitimate.

Data Quality Insight :
The linear clusters prove that Estimated Revenue is a constructed variable, not a real-world measurement.
Any predictive model using this feature will simply “rediscover” the formulas baked into the dataset.

Recommendation :
The structure of the Estimated Revenue feature suggests formula-based generation. Before relying on this variable for modeling or business conclusions, the data source and revenue estimation logic should be reviewed.

```python
#Check Multi-Collinearity Using VIF (Variance Inflation Factor)
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
numeric_df = df.select_dtypes(include=['float64', 'int64'])
from statsmodels.tools.tools import add_constant
X = add_constant(numeric_df)
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_data
```

The dataset shows mostly acceptable multicollinearity, with most variables falling under the safe range (VIF < 5).

Three variables display meaningful collinearity:

- Delivery_time (6.54)
- Total_ratings (8.58)
- Estimated_revenue (8.84)

One major issue was detected:
- Satisfaction_score has high multicollinearity (VIF > 10) and may need special handling.

The presence of high VIF for Satisfaction_score and Estimated_revenue confirms earlier findings:
- Estimated_revenue is a derived feature (not raw data).
- Satisfaction_score is heavily tied to delivery speed.

For predictive modeling, high-VIF variables will lead to unstable coefficients. These variables may need:
- Removal
- Transformation
- Regularization methods (like Ridge/Lasso).

For exploratory analysis, high multicollinearity is not fatal, but it must be acknowledged as a limitation.

```python
#PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

numeric_df = df.select_dtypes(include=['float64', 'int64'])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)
pca = PCA()
pca_data = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(pca_data,columns=[f'PC{i+1}' for i in range(pca_data.shape[1])])
explained_variance = pca.explained_variance_ratio_
explained_variance
```

The first principal component (PC1) explains ~32.66% of all variation in the dataset.
- This means PC1 captures the biggest pattern across the features.

The second component (PC2) explains ~21.44%.
- Together, PC1 + PC2 explain ~54.1% of the total information.

The third (15.1%) and fourth (11.9%) components push the total explained variance to ~81.1%.
- This means the first 4 components capture most of the meaningful structure of the data.

The remaining components (PC5–PC8) individually contribute very little (7%, 0.7%, 0.6%).
- These components mostly contain noise or very minor patterns.

Based on standard practice, selecting the first 4 components would be sufficient to retain 80%+ of the dataset’s information while reducing dimensionality.

PCA confirms that the dataset has moderate multicollinearity, but it’s manageable and compressible into fewer underlying factors.

```python
#Cumulative Explained Variance Plot.
plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

- The cumulative explained variance curve shows that the first four principal components capture over 81% of the total variance in the dataset.
- After the 4th component, the curve starts to flatten significantly, indicating diminishing returns from adding more components.
- PC1 alone explains around 33%, and PC1 + PC2 explain roughly 54%, meaning the top two components capture more than half of the dataset’s structure.
- The elbow point occurs at component 4, which is the optimal cutoff for dimensionality reduction. Beyond this point, additional components add very little new information and mostly represent noise.
- Using 4 components would preserve most meaningful information while reducing redundancy and multicollinearity across the original features.

```python
#PCA Loadings Analysis – Identifying Feature Contribution to Principal Components.
feature_cols = ['ID', 'Price', 'Avg_ratings', 'Total_ratings', 'Delivery_time', 'Estimated_revenue', 'Satisfaction_score', 'Cuisine_count']
loadings = pd.DataFrame(pca.components_.T,columns=[f"PC{i+1}" for i in range(len(pca.components_))],index=feature_cols)
print("\nPCA Loadings (Feature Contribution to Each Component):\n")
print(loadings)
plt.figure(figsize=(10,6))
sns.heatmap(loadings.iloc[:, :4], annot=True, cmap='coolwarm')
plt.title("PCA Loadings Heatmap (Top 4 Components)")
plt.show()
```
